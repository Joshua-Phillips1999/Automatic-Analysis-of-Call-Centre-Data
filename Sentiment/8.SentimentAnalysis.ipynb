```python
import re

def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    return [ atoi(c) for c in re.split(r'(\d+)', text) ]
```

```python
for i in file_dict.keys():
  sorted_files = dict()
  val2 = list(file_dict[i].keys())
  val2.sort(key = natural_keys)       #val2 has list of sorted file names. Check in future if this works for all kinds of 
  for j in val2:
    sorted_files[j] = file_dict[i][j]
  file_dict[i] = sorted_files
print(file_dict)
f = open('/content/drive/MyDrive/CAPSTONE/Colab Notebooks/file_dict','w')
f.write(str(file_dict))
f.close()
```

HUGGING FACE

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import torch.nn.functional as F
```

```python
tokenizer = AutoTokenizer.from_pretrained("finiteautomata/bertweet-base-sentiment-analysis")
model = AutoModelForSequenceClassification.from_pretrained("finiteautomata/bertweet-base-sentiment-analysis")
```

```python
#Hugging Face
sentiment_dict = dict()
j = 0
for i in df['text']:
  try: 
    x = i.replace('\n',' ')
    inputs = tokenizer(x,return_tensors="pt")
    outputs = model(**inputs)
    pt_predictions = F.softmax(outputs.logits, dim=-1)
    #print(pt_predictions)
    a=torch.argmax(pt_predictions, dim = -1)
    x=pt_predictions[0][0].clone().detach().requires_grad_(True)
    y=pt_predictions[0][1].clone().detach().requires_grad_(True)
    z=pt_predictions[0][2].clone().detach().requires_grad_(True)
    print(str(x.item()*100)+','+str(y.item()*100)+','+str(z.item()*100))
    if(str(a)=='tensor([0])'):
      val2='negative'
    if(str(a)=='tensor([1])'):
      val2='neutral'
    if(str(a)=='tensor([2])'):
      val2='positive'
    sentiment_dict[i] = val2
  except IndexError:
    pass
```

```python
# Number of text lines processed by Bertweet base sentiment model
print(len(sentiment_dict))
```

```python
#Accuracy of HuggingFace's Bertweet Base Sentiment Model
count = 0
for i, row in df.iterrows():
    t = row['text']
    if(t in sentiment_dict.keys()):
      actuals = df.at[i,'sentiment']
      if(sentiment_dict[t] == actuals):
        count+=1
print(count*100/len(sentiment_dict))
```

VADER

```python
#VADER

import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
'''
The compound score is the sum of positive, negative & neutral scores which is then normalized between -1(most extreme negative) and +1 (most extreme positive).

The more Compound score closer to +1, the higher the positivity of the text.
pos = 0.33 to 1
neu = -0.33 to 0.33
neg = -1 to -0.33
'''
sentiment_dict_vader = dict()
for i in df['text']:
    scores = sid.polarity_scores(i)
    #scores.pop('compound')
    cm = scores['compound']
    if(cm<=1 and cm>0.33):
      score = 'positive'
    if(cm<=0.33 and cm>-0.33):
      score = 'neutral'
    if(cm<=-0.33 and cm>-1):
      score = 'negative'
    sentiment_dict_vader[i] = score
    print(scores,i,sentiment_dict_vader[i])
```

```python
#Accuracy of VADER
count = 0
for i, row in df.iterrows():
    t = row['text']
    actuals = df.at[i,'sentiment']
    #print(actuals,sentiment_dict_vader[t])
    if(sentiment_dict_vader[t] == actuals):
      count+=1
print(count)
print(count*100/len(sentiment_dict_vader))
```
