```python
import librosa
import torch
import os
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

```python
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-large-960h-lv60-self")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-large-960h-lv60-self")
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
#In the below way the transcript is saved in all different files such as 1_0.txt, 1_1.txt rather than having it in one file known as 1_.txt

#reading the files from drive
directory = r'/content/drive/MyDrive/CAPSTONE/Philippineexample/4_'
counter=0
for filename in os.listdir(directory):
      #To take in only the whiles ending with .wav
      if filename.endswith(".wav") :
          
          audio, rate = librosa.load(os.path.join(directory, filename), sr = 16000)
          input_values = tokenizer(audio, return_tensors = "pt").input_values
          logits = model(input_values).logits
          prediction = torch.argmax(logits, dim = -1)
          transcription = tokenizer.batch_decode(prediction)[0]
          transcription = transcription.lower()
          filess="/content/drive/MyDrive/CAPSTONE/Philippineexample/Transcript/4_"+str(counter)+".txt"
          f = open(filess, "a") 
          counter+=1
          f.write(filename+','+transcription+'\n')
          print(filename+','+transcription+'\n')
          f.close()
```
