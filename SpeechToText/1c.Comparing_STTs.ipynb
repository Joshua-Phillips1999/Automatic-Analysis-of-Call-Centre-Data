##VOSK


```python
#VOSK
from jiwer import wer
import jiwer

original = open('/content/original_text.txt','r')
transcribed = open('/content/transcribed_vosk.txt','r')

werlist = []
merlist = []
willist = []

o = dict()
t = dict()

for i in original: 
  l = i.split(",")
  o[l[0]] = l[1].lower()

for j in transcribed:
  l = j.split(",")
  t[l[0]] = l[1]

print(len(o))
print(len(t))

for i in o.keys():
  if(i in t.keys()):
    measures = jiwer.compute_measures(o[i],t[i])
    werlist.append(measures['wer'])
    merlist.append(measures['mer'])
    willist.append(measures['wil'])

print(len(werlist),len(merlist),len(willist))
print("For VOSK")
print("Average WER: "+str(sum(werlist)/len(werlist)))
print("Average MER: "+str(sum(merlist)/len(merlist)))
print("Average WIL: "+str(sum(willist)/len(willist)))
```

```python
#vosk indian
f1=open("/content/transcribed_in_new.txt","r")
f2=open("/content/transcribed_in_new2.txt","w")
data=(f1).readlines()
data.sort()
for i in range(len(data)):
    l=data[i].split(",")
    l[0]=l[0][:-4]
    for j in l[1]:
      if j=='\'':
        l[1]=l[1].replace('\'','')
    f2.write(l[0]+","+l[1])
f1.close()
f2.close()
```
