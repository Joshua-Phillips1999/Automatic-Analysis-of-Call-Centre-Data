```python
import librosa
import torch
import os
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

```python
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-large-960h-lv60-self")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-large-960h-lv60-self")
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
#1
directory = r'/content/drive/MyDrive/CAPSTONE/VoiceFilesSet1/type1/'
for filename in os.listdir(directory):
    #if filename not in arr:
      if filename.endswith(".wav") :
          #print(os.path.join(directory, filename))
          audio, rate = librosa.load(os.path.join(directory, filename), sr = 16000)
          input_values = tokenizer(audio, return_tensors = "pt").input_values
          logits = model(input_values).logits
          prediction = torch.argmax(logits, dim = -1)
          transcription = tokenizer.batch_decode(prediction)[0]
          transcription = transcription.lower()
          f = open("/content/outputWav2VecType1.txt", "a") 
          f.write(filename+','+transcription+'\n')
          print(filename+','+transcription+'\n')
          f.close()
```

```python
#Wav2VecType1
!pip install jiwer
from jiwer import wer
import jiwer

original = open('/content/drive/MyDrive/CAPSTONE/VoiceFilesSet1/type1/type1_trans.txt','r')
transcribed = open('/content/drive/MyDrive/CAPSTONE/VoiceFilesSet1/wav2vec_Trans/outputWav2VecType1.txt','r')

werlist = []
merlist = []
willist = []
l=[]

for i in original: 
  o = i.lower().split("\n")
#   o[l[0]] = l[1].lower()

for j in transcribed:
  l.append(j.split(",")[1]) 
  #t[l[0]] = l[1]

print(len(o))
print(len(l))
print(o)
for i in l:
  print(i)

for i in o:
  for j in l:
    measures = jiwer.compute_measures(i,j)
    werlist.append(measures['wer'])
    merlist.append(measures['mer'])
    willist.append(measures['wil'])

print(len(werlist),len(merlist),len(willist))
print("For Type1")
print("WER"+str(werlist))
print("Average WER: "+str(sum(werlist)/len(werlist)))
print("Average MER: "+str(sum(merlist)/len(merlist)))
print("Average WIL: "+str(sum(willist)/len(willist)))
```
